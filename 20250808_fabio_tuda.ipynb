```python
# Question 7: Create a ranking of airlines based on the number of flights they operate.
```

```python
# Step 1: Count occurrences
count_flights = df_ewr['flight'].value_counts().reset_index()
count_flights.columns = ['flight', 'Count']

# # Step 2: Apply rank on the counts
count_flights['Rank'] = count_flights['Count'].rank(method='dense', ascending=False)

print(display(count_flights))
```
